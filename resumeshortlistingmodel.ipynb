```python
#Reading csv file
import pandas as pd
dataset=pd.read_csv('harish_paccount.csv')
dataset
```

```python
#Null  count
print(" \nCount total NaN at each column in a DataFrame : \n\n", 
      dataset.isnull().sum())
```

```python
#Dataset Info
dataset.info()
```

```python
#Rename columns
dataset = dataset.rename({'requirement': 'target','10th%':'tenthscore','12th%':'twelfthscore'},axis=1)
```

```python
#Choosing the Predictor variable or independent variable
predictor=dataset[['cgpa','skills','certification','project','internorwexp','ppublication','tenthscore','twelfthscore','age']]
predictor
```

```python
#choosing the target variable or dependent variable
y=dataset[['target']]
y
```

```python
#encoding the matched values to 1 else 0
y=y["target"].apply(lambda target:1 if target=='matched' else 0)
y
```

```python
#Choosing the required skills
predictor["skills"]=predictor["skills"].apply(lambda skills:skills if skills=="java" or skills=='python' or skills=='angular' or skills=='c++' or skills=='php' or skills=='sql' or skills=='mongodb' or skills=='reactjs' or skills=='html' or skills=='mysql' else 0)

#Extracting  the skills and project matched values
import numpy as np
predictor['skillprojmatch'] = np.where(predictor['skills'] == predictor['project'], 1, 0)
predictor
```

```python
#Manual Binary convertion

#predictor["skills"]=predictor["skills"].apply(lambda skills:1 if skills=="java" or skills=='python' or skills=='angular' or skills=='c++' or skills=='php' or skills=='sql' or skills=='mongodb' or skills=='reactjs' or skills=='html' or skills=='mysql' else 0)
predictor["certification"]=predictor["certification"].apply(lambda certification:1 if certification=='yes' else 0)
predictor["internorwexp"]=predictor["internorwexp"].apply(lambda internorwexp:1 if internorwexp=='yes' else 0)
predictor["ppublication"]=predictor["ppublication"].apply(lambda ppublication:1 if ppublication=='yes' else 0)
#predictor["project"]=predictor["project"].apply(lambda project:1 if project=="java" or project=='python' or project=='angular' or project=='c++' or project=='php' or project=='sql' or project=='mongodb' or project=='reactjs' or project=='html' or project=='mysql' else 0)
predictor["tenthscore"]=predictor["tenthscore"].apply(lambda tenthscore:1 if tenthscore>59 else 0)
predictor["twelfthscore"]=predictor["twelfthscore"].apply(lambda twelfthscore:1 if twelfthscore>59 else 0)
predictor["age"]=predictor["age"].apply(lambda age:1 if age==24 or age==23 else 0)
predictor["cgpa"]=predictor["cgpa"].apply(lambda cgpa:1 if cgpa>=7.0 else 0)
```

```python
#Filtering out the eligible category students

predictor=pd.concat([predictor,y],axis=1)
predictor = predictor[predictor['cgpa'] == 1]
predictor=predictor[predictor['tenthscore']==1]
predictor=predictor[predictor['twelfthscore']==1]
predictor=predictor[predictor['age']==1]
predictor
```

```python
#finalizing the predictor variable by dropping unnecessary values

y=predictor['target']
predictor=predictor.drop(['skills','project','age','target'],axis=1)
print(predictor)
print(y)
```

```python
#splitting dataset into Train and Test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(predictor,y,test_size=0.3,random_state=15)
```

```python
#call the Logistic regression model
from sklearn.linear_model import LogisticRegression
logit=LogisticRegression()
logit.fit(x_train,y_train)
```

```python
#computing the predictions
y_pred=logit.predict(x_test)
```

```python
# LOGISTIC REGRESSION confusion matrix
from sklearn.metrics import confusion_matrix
cf=confusion_matrix(y_test,y_pred)
cf
```

```python
# LOGISTIC REGRESSION accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# LOGISTIC REGRESSION classification report
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```

```python
#NAIVE BAYES model
from sklearn.naive_bayes import GaussianNB
gn=GaussianNB()
gn.fit(x_train,y_train)
```

```python
#Testing the model with single test value
ypredt=gn.predict([[1,0,1,0,1,1,0]])
if ypredt[0]==1:
    print("shortlisted")
else:
    print("not shortlisted")
```

```python
ypred=gn.predict(x_test)
```

```python
#NAIVE BAYES confusion matrix
from sklearn.metrics import confusion_matrix
cf1=confusion_matrix(y_test,ypred)
cf1
```

```python
# NAIVE BAYES accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,ypred)
```

```python
#NAIVE BAYES classification report
from sklearn.metrics import classification_report
report1=classification_report(y_test,ypred)
print(report1)
```
